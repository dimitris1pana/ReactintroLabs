```python
from scipy.stats import f_oneway

A = [10, 12, 11, 9, 15]
B = [14, 15, 13, 10, 18]
C = [20, 22, 19, 21, 23]

F, p = f_oneway(A, B, C)

# Interpretation rule
alpha = 0.05
F_value, p_value = f_oneway(A, B, C)

print("One-Way ANOVA Results:")
print(f"F-value: {F_value:.3f}")
print(f"p-value: {p_value:.5f}")

if p_value < alpha:
    print("\nResult: Statistically significant (p < 0.05).")
    print("At least one group mean differs from the others.")
    print("You MUST run post-hoc tests to determine *which* groups differ.")
else:
    print("\nResult: NOT statistically significant (p ≥ 0.05).")
    print("No evidence that any group means are different.")
    print("No post-hoc tests are needed.")
```

- F-value → ratio of between-group variance to within-group variance
- p-value → probability the group means are equal (null hypothesis)
